```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

```python
def compute_normal_distribution(x, mean, std):
    y = 1 / (std*np.sqrt(2*np.pi))
    y = y * np.exp(-(1/2)*((x - mean)/std)**2)

    return y
```

```python
step = 0.1 # dx
x = np.arange(-10, 10, step)
y = compute_normal_distribution(x, 0, 1)
y_2 = compute_normal_distribution(x, 2, 1.5)

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.plot(x, y);
plt.subplot(122)
plt.plot(x, y_2);
```

```python
## Area under the curve is the integral
# Which is equal to the sum of all vertical (y) values multiplied by the 
# difference in x (step, dx), which is 0.1 in this example
area = y.sum() * step
# (y * step).sum() to visualize it ase a sum of rectangle areas

print(f'Area of normal distribution = {area:.2f}')
```

```python
def compute_z_score(x, mean, std):
    """
    Standardize x (how many standard deviations is it away from the mean?)
    """
    z = (x - mean) / std

    return z
```

```python
z = compute_z_score(x, 2, 1.5)

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.title("normal distribution, mean=2, std=1.5")
plt.xlabel('x')
plt.ylabel('density')
plt.plot(x, y_2);

plt.subplot(122)
plt.title("standardized with z-score")
plt.xlabel('z-score (number of standard deviations from the mean)')
plt.plot(z, y_2);
```

## Z score with the sample means

https://math.stackexchange.com/questions/845769/why-does-the-normalized-z-score-introduce-a-square-root-and-some-more-confusio

 

```python
def compute_z_score_means(x_mean, n_samples, mean, std):
    """
    z score for mean inputs, so we cant do hypothesis testing with real mean from n_samples
    """
    z = (x_mean - mean) / (std / np.sqrt(n_samples))

    return z
```

```python
x_means = np.arange(-3, 3, 0.1)
z_means = compute_z_score_means(x_means, len(x), 2, 1.5)


plt.figure(figsize=(15,5))
plt.subplot(131)
plt.title("normal distribution")
plt.xlabel('x')
plt.ylabel('density')
plt.plot(x, y_2);

plt.subplot(132)
plt.title("z-score")
plt.xlabel('z-score (number of standard deviations from the mean)')
plt.plot(z, y_2);

plt.subplot(133)
plt.title("z-score means")
plt.xlabel('x means')
plt.ylabel('z-score (number of standard deviations from the mean)')
plt.plot(x_means, z_means);
```
